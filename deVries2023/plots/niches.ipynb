```python
import matplotlib.pyplot as plt
import xarray as xr
import seaborn as sns
import numpy as np


ds = xr.open_dataset("/home/phyto/ModelOutput/merged/abundances.nc")
ds = ds[['temperature', 'shannon', "total", "DIC", "din", "o2", "irradiance", 'Emiliania huxleyi']]
d = ds.to_dataframe()
ds = None
d= d.dropna()
d['total minus ehux'] = d['total'] - d['Emiliania huxleyi']

sns.kdeplot(data=d, weights="total minus ehux", x="DIC", color="orange").set_xlim(1900, 2250)
sns.kdeplot(data=d, weights="Emiliania huxleyi", x="DIC", color="blue").set_xlim(1900, 2250)

plt.show()
```

```python
sns.kdeplot(data=d, weights="total minus ehux", x="temperature", color="orange")
sns.kdeplot(data=d, weights="Emiliania huxleyi", x="temperature", color="blue")

plt.show()
```

```python
import numpy as np
d["log irradiance"] = np.log(d['irradiance'])

sns.kdeplot(data=d, weights="total minus ehux", x="irradiance", color="orange").set_xlim(None, 30)
sns.kdeplot(data=d, weights="Emiliania huxleyi", x="irradiance", color="blue").set_xlim(None, 30)

plt.show()
```

```python
sns.kdeplot(data=d, weights="total minus ehux", x="o2", color="orange")
sns.kdeplot(data=d, weights="Emiliania huxleyi", x="o2", color="blue")

plt.show()
```

```python
sns.kdeplot(data=d, weights="total minus ehux", x="din", color="orange").set_xlim(None, 30)
sns.kdeplot(data=d, weights="Emiliania huxleyi", x="din", color="blue").set_xlim(None, 30)

plt.show()
```
